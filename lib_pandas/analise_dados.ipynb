```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_vendas = {
    'Produtos'   : ["Camisa", 'Calça', "Saia", "Tenis", "Camisa", "Blusa", 'Casaco', "Camiseta", "Calça", "Tenis"],
    'Preço'      : [20, 30, 56, 49, 85, 60, 45, 100, 99, 32],
    'Quantidade' : [10, 5, 6, 30, 20, 12, 14, 50, 10, 30],
    'Mês'        : ["Jan", "Jan", "Jan", "Abr", "Abr", "Mai", "Mai", "Mai", "Jun", "Jun"]
}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

```python
df_vendas["Total (em R$)"] = df_vendas["Preço"] * df_vendas["Quantidade"]
df_vendas
```

```python
resumo_vendas = df_vendas.groupby("Produtos")["Total (em R$)"].sum().reset_index()
resumo_vendas
```

```python
plt.figure(figsize=(8, 5))
plt.bar(resumo_vendas["Produtos"], resumo_vendas["Total (em R$)"], color="black")
plt.title("Vendas total por Produtos")
plt.xlabel("Produtos")
plt.ylabel("Total (em R$)")
plt.xticks(rotation=45)
```
